```python
import pandas as pd
from sqlalchemy import create_engine

# No more errors on this line!
engine = create_engine('postgresql+psycopg2://postgres:#Bethely2005@localhost/app_reviews')

def run(query):
    return pd.read_sql(query, engine)

# Load and upload
df = pd.read_csv("amazon.csv")
df.columns = df.columns.str.lower()

df.to_sql("amazon", engine, if_exists="replace", index=False)
```

```python
#review length
run("""
SELECT positive,
       AVG(LENGTH(reviewtext)) AS avg_review_length
FROM amazon
GROUP BY positive;
""")#negative reviews are longer than positive reviews
```

```python
negative_reviews = run(" SELECT reviewtext, LENGTH(reviewtext) AS review_length FROM amazon WHERE positive = 0 ")
negative_reviews["reviewtext"] = negative_reviews["reviewtext"].str.lower()
negative_reviews["reviewtext"] = negative_reviews["reviewtext"].str.replace(r"[^\w\s]", "", regex=True)
words = negative_reviews["reviewtext"].str.split()
all_words = words.explode()
word_counts = all_words.value_counts().head(20)
word_counts
```

```python
stopwords = [
"i","me","my","myself","we","our","ours","ourselves",
"you","your","yours","yourself","yourselves",
"he","him","his","himself","she","her","hers","herself",
"it","its","itself","they","them","their","theirs","themselves",
"what","which","who","whom","this","that","these","those",
"am","is","are","was","were","be","been","being",
"have","has","had","having",
"do","does","did","doing",
"a","an","the",
"and","but","if","or","because","as","until","while",
"of","at","by","for","with","about","against","between","into",
"through","during","before","after","above","below","to","from",
"up","down","in","out","on","off","over","under",
"again","further","then","once"
]
filtered_words = all_words[~all_words.isin(stopwords)]
word_counts = filtered_words.value_counts().head(20)
print(word_counts)
```

```python
from itertools import tee

def bigrams(words):
    a, b = tee(words)
    next(b, None)
    return zip(a, b)

pairs = list(bigrams(filtered_words))

bigram_series = pd.Series([" ".join(pair) for pair in pairs])

bigram_counts = bigram_series.value_counts().head(20)

print(bigram_counts)
```

```python
import matplotlib.pyplot as plt

bigram_counts.head(10).sort_values().plot(kind="barh")

plt.title("Top Complaint Phrases in Negative Reviews")
plt.xlabel("Frequency")
plt.ylabel("Phrase")

plt.show()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

df["sentiment_score"] = df["reviewtext"].apply(
    lambda x: analyzer.polarity_scores(str(x))["compound"]
)

df.groupby("positive")["sentiment_score"].mean()
```

```python
df["sentiment_score"].describe()
```

```python
import matplotlib.pyplot as plt

df["sentiment_score"].hist(bins=30)

plt.title("Sentiment Score Distribution")
plt.xlabel("Sentiment Score")
plt.ylabel("Number of Reviews")

plt.show()
```

```python
#suspicious reviews
mismatch = df[
    ((df["positive"] == 1) & (df["sentiment_score"] < 0)) |
    ((df["positive"] == 0) & (df["sentiment_score"] > 0.5))
]

len(mismatch)
```
